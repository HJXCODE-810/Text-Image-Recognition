# Handwritten text recognition using KNN model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,classification_report,confusion_matrix
```

```python
data_df = pd.read_csv("data.csv")
test_df = pd.read_csv("test.csv")
```

```python
data_df.head()
```

```python
test_df.head()
```

```python
y = data_df['label'] #target
x = data_df.drop('label',axis=1)#features
```

# Using Train:test ratio : 60:40

```python
k = [2,4,5,6,7,10]
Test_ratio = [0.40,0.30,0.25,0.20,0.10,0.05]
for i in k:
    for j in Test_ratio:
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = j)
        from sklearn.preprocessing import  StandardScaler
        scaler = StandardScaler()
        scaler.fit(x_train,y_train)
        x_train = scaler.transform(x_train) 
        
        classifier = KNeighborsClassifier(n_neighbors = i)
        classifier.fit(x_train,y_train)

        y_pred = classifier.predict(x_test)
        y_pred
        #y_pred_on_test_data = classifier.predict(x_for_test_data)
        #y_pred_on_test_data

        print("The Accuracy score is:",accuracy_score(y_test, y_pred))
        print("\n")

        print("The Classification report:\n",classification_report(y_test, y_pred))
        print("\n")

        print("The Confusion matrix is:\n",confusion_matrix(y_test, y_pred))

```
